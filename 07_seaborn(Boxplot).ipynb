## Seaborn Boxplot 

A box plot is used to visualize the **distribution of numerical data**    
through quartiles and identify **outliers**.

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

## Load Dataset 

```python
tips=sns.load_dataset('tips')
tips.head()
```

### Basic Boxplot on total_bill

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

### Box Plot With hue 

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
plt.title("Day vs Total-bill")
plt.show()
```

### Horizontal Box Plot

```python
sns.boxplot(y='day',x='total_bill',data=tips)
plt.show()
```

### Boxplot with Norch True 
Notch in a Seaborn box plot represents the confidence interval around the median and helps compare medians across groups.

```python
sns.boxplot(x='day',y='total_bill',data=tips,notch=True)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,fill=True)
plt.show()
```
