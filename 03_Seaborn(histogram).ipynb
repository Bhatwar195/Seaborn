## Seaborn Histogram 
A **Histogram** is a data visualization used to represent the distribution of a numerical variable by    
dividing the data into bins and displaying the frequency of values within each bin.

### Import Required Libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Load tips dataset

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

### Basic Histogram

```python
sns.histplot(tips["total_bill"], bins=10)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

### Histogram with KDE

```python
sns.histplot(tips['total_bill'],kde=True)
plt.title("Total Bill Distribution with KDE")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

### Histogram with Hue

```python
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    bins=10,
    multiple="stack"
)
plt.title("Total Bill Distribution by Gender")
plt.show()
```
