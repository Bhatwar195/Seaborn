## Seaborn Count Plot 

Seaborn count plot is a visualization used to show the frequency (count) of observations    
in each category of a categorical variable. It displays bars whose heights represent how many times   
each category appears in the dataset.   

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Load Tips Dataset 

```python
tips=sns.load_dataset('tips')
tips.head()
```

### Basic Count Plot 

```python
sns.countplot(x='day',data=tips)
plt.show()
```

### Count Plot With hue 

```python
sns.countplot(x='day',data=tips,hue='sex')
plt.show()
```

### Count Plot On Days 

```python
sns.countplot(y='day',data=tips)
plt.show()
```

### Count Plot with Hue Smoker 

```python
sns.countplot(x='day',data=tips,hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex',data=tips,hue='sex',palette='Set2')
```
