### Seaborn Lm plot (Linear Model Plot)

* A **LM Plot** (Linear Model Plot) in Seaborn is used to visualize the linear relationship between two numerical variables by plotting
data points along with a best-fit regression line.

--------------------------------------------------------------------------------------------------------------

### Import Required Libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

### Load Titanic Dataset 

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

### Ploting Simple Line plot 

```python
sns.lineplot(x='age',y='survived',hue='sex',data=titanic)
plt.title("Survival rate vs age")
plt.xlabel("Age")
plt.ylabel("Survival rate")
plt.show()
```

### Basic LM Plot

```python
sns.lmplot(x='age',y='survived',hue='sex',data=titanic,height=5,aspect=1.5,lowess=True)
plt.title("Survival rate vs age")
plt.xlabel("Age")
plt.ylabel("Survival rate")
plt.show()
```

```python
sns.lmplot(x='age',y='survived',hue='pclass',data=titanic,height=5,aspect=1.5,lowess=True)
plt.title("Survival rate vs age (By passenger class)")
plt.xlabel("Age")
plt.ylabel("Survival rate")
plt.show()
```

### multi condition smooth line plot (Sex +class)

```python
sns.lmplot(x='age',y='survived',hue='sex',col='pclass',lowess=True,data=titanic,height=5,aspect=1.5)
plt.subplots_adjust(top=0.85)
plt.title("Survival rate vs age (By sex & class)",fontsize=14)
plt.show()
```

```python
titanic['group']=titanic['sex']+titanic['pclass'].astype(str)
```

```python
sns.lmplot(x='age',y='survived',hue='group',data=titanic,lowess=True,height=5,aspect=1.5)
plt.title("Survival rate vs age (By Sex+class)")
plt.xlabel("Age")
plt.ylabel("Survival rate")
plt.show()
```
