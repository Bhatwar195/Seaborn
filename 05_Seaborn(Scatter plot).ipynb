## Seaborn Scatter Plot 

Scatter Plot-  
A scatter plot is used to observe the **relationship, correlation, or pattern**  
between two continuous variables.

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Load Dataset 

```python
tips=sns.load_dataset('tips')
tips.head()
```

### Basic Scatter Plot

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

### Scatter Plot with Hue

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
plt.show()
```

### Scatter Plot with Size 

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,size='size')
plt.show()
```

### Scatter Plot with hue,style and size 

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,size='size',hue='day',style='sex')
plt.show()
```
