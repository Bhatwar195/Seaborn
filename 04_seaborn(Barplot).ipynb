## Seaborn Bar Plot
A bar plot is used to compare **average values** of a numerical variable
across different categories.

### Import Required Libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Load Dataset 

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

### Basic Bar Plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

### Bar Plot with Hue

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',estimator=sum,data=tips)
plt.show()
```

### Bar Plot with Confidence Interval

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',ci=None)
plt.show()
```

### Confidence Interval
Seaborn bar plots display a confidence interval by default,
representing data variability.

```python
sns.barplot(y='day',x='total_bill',data=tips,hue='sex')
plt.show()
```
